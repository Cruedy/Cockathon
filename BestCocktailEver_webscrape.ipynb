```python
import requests
from bs4 import BeautifulSoup
import string 
```

```python
def synonym(keyword): # Returns top 10 synonyms of a given keyword, may return antonyms if fewer than 10 synonyms
  file = requests.get("https://thesaurus.yourdictionary.com/" + keyword)
  soup = BeautifulSoup(file.content, "html.parser")
  divs = soup.find_all('a', {"class": "text-base text-black underline sm:no-underline hover:underline"})
  for x in range(10):
    print(divs[x].get("href"))
  
```

```python
letters = list(string.ascii_lowercase) 
urls = []
names = []
ingredients = []
for l in letters:
  file = requests.get("https://www.thecocktaildb.com/browse.php?b=" + l)
  soup = BeautifulSoup(file.content, "html.parser")
  all = soup.find("section", {"id":"feature"})
  content = all.find_all("div", {"class":"col-sm-3"})
  for link in soup.find_all('a'):
    urls.append(link.get('href'))
  urls = urls[6:len(urls) - 33]
urls = urls[1::2]
```

```python
name_ing = {}
for u in urls:
  ingreds = []
  name = ""
  file_1 = requests.get("https://www.thecocktaildb.com/" + u)
  soup = BeautifulSoup(file_1.content, "html.parser")
  all = soup.find("div", {"class":"center"})
  if all != None:
    name = all.find("h2").get_text()
    name = name.replace("\n", "")
    name = name.replace("\t", "")
    in_all = all.findAll("td")
    ing = []
    for i in in_all:
      ingred = i.findAll("figcaption")
      for g in ingred:
        ing.append(g.get_text())
    split = len(ing)//2
    ing = ing[:split]
    name_ing[name] = ing
print(name_ing)
      
```

```python
name_tags = {}
for u in urls:
  name = ""
  file_1 = requests.get("https://www.thecocktaildb.com/" + u)
  soup = BeautifulSoup(file_1.content, "html.parser")
  all = soup.find("div", {"class":"center"})
  if all != None:
    name = all.find("h2").get_text()
    name = name.replace("\n", "")
    name = name.replace("\t", "")
    table = all.find("table")
    ts = table.findAll("div", {"class": "textwidget"})
    tags = []
    for t in ts:
      desc = t.findAll("a", {"class": "tag"})
      for d in desc:
        tags.append(d.get_text().lower())
    if tags != []:
      name_tags[name] = tags
```

```python
import csv
ing_title = {"Name": "Ingredients"}
ing_title.update(name_ing)
name_ing = ing_title
tag_title = {"Name": "Tags"}
tag_title.update(name_tags)
name_tags = tag_title
with open('templates/helper/Ingredients.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in name_ing.items():
       writer.writerow([key, value])
with open('templates/helper/Tags.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in name_tags.items():
       writer.writerow([key, value])
```

```python
from itertools import chain
keyw = list(name_tags.values())
keywords = list(chain.from_iterable(keyw[1:]))
word_freq = {}
# print(keywords)
for k in keywords:
  k = k.replace(" ", "")
  if k.lower() in word_freq:
            word_freq[k.lower()] += 1
  else:
    word_freq[k.lower()] = 1
new_dict = {}
for key, value in word_freq.items():
   if value in new_dict:
       new_dict[value].append(key)
   else:
       new_dict[value]=[key]
word_freq = new_dict
myKeys = list(word_freq.keys())
myKeys.sort()
myKeys.reverse()
word_freq = {i: word_freq[i] for i in myKeys}
print(word_freq)
```

```python
num_words = {"Number": "Words"}
num_words.update(word_freq)
word_freq = num_words
with open('templates/helper/Keywords.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in word_freq.items():
       writer.writerow([key, value])
```

```python
def getDrinks(tags):
  drinks = []
  for t in tags:
    per_tag = []
    with open("templates/helper/Tags.csv", "r") as file:
      csvreader = csv.reader(file)
      for row in csvreader:
        if t in row[1]:
          per_tag.append(row[0])
    drinks.append(per_tag)
  return drinks
getDrinks(["fruity"])
```
